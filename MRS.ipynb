```python

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
movies_data = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['title','tagline','status','overview']
print(selected_features)
```

```python
# Fill missing values in selected features
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')

# Concatenate selected columns and store the result in a new column 'combined_features'
movies_data['combined_features'] = (
    movies_data['title'] + ' ' +
    movies_data['tagline'] + ' ' +
    movies_data['status'] + ' ' +
    movies_data['overview'] 
)
```

```python
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')

combined_features = movies_data['title'] +' '+  movies_data['tagline'] +' '+ movies_data['status'] +''+ movies_data['overview']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

COSINE SIMILARITY

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_movie = movies_data[movies_data['title'] == 'Spider-Man 3'].index[0]
print("Index of", 'Spider-Man 3', ":", index_of_movie)
```

```python
# getting a list of similar movies

similarity_scores = list(enumerate(similarity[index_of_movie]))
print(similarity_scores)
```

```python
# For computing similarity_scores

length_of_similarity_scores = len(similarity_scores)
print("Number of similar movies:", length_of_similarity_scores)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

MOVIE RECOMMENDATION SYSTEM

```python
movie_name = input('Enter your favourite movie name: ')
list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match].index[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)

print('Movies suggested for you:\n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data.loc[index, 'title']
    if i < 30:
        print(i, '.', title_from_index)
        i += 1







```
